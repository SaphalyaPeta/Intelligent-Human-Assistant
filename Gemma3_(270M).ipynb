### Data Prep
We now use the `Gemma-3` format for conversation style finetunes. We use [Thytu's ChessInstruct](https://huggingface.co/datasets/Thytu/ChessInstruct) dataset. Gemma-3 renders multi turn conversations like below:

```
<bos><start_of_turn>user
Hello!<end_of_turn>
<start_of_turn>model
Hey there!<end_of_turn>
```

We use our `get_chat_template` function to get the correct chat template. We support `zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, phi3, llama3, phi4, qwen2.5, gemma3` and more.

````python
from unsloth.chat_templates import get_chat_template
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma3",
)
````

````python
from datasets import load_dataset
dataset = load_dataset("json", data_files="voice_commands_converted.jsonl", split="train")
````

We now use `convert_to_chatml` to try converting datasets to the correct format for finetuning purposes!

````python
def convert_to_chatml(example):
    return {
        "conversations": [
            {"role": "system", "content": example["task"]},
            {"role": "user", "content": example["input"]},
            {"role": "assistant", "content": example["expected_output"]}
        ]
    }

dataset = dataset.map(
    convert_to_chatml
)
````

Let's see how row 100 looks like!

````python
dataset[100]
````

We now have to apply the chat template for `Gemma3` onto the conversations, and save it to `text`.

````python
def formatting_prompts_func(examples):
    texts = []
    for conversation in examples["conversations"]:
        formatted_text = ""
        for turn in conversation:
            if turn["role"] == "system":
                formatted_text += f"<start_of_turn>{turn['role']}\n{turn['content']}<end_of_turn>\n"
            elif turn["role"] == "user":
                formatted_text += f"<start_of_turn>{turn['role']}\n{turn['content']}<end_of_turn>\n"
            elif turn["role"] == "assistant":
                formatted_text += f"<start_of_turn>{turn['role']}\n{turn['content']}<end_of_turn>\n"
        texts.append(formatted_text)
    return {"text": texts}
````

Let's see how the chat template did!


````python
dataset = dataset.map(formatting_prompts_func, batched=True)

print(dataset[100]['text'])
````

### Train the model
Now let's train our model. We do 100 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`.

````python
from transformers import DataCollatorForLanguageModeling
````

````python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    eval_dataset = None,
    args = SFTConfig(
        dataset_text_field = "text",
        per_device_train_batch_size = 2,  # Reduced for memory
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 200,
        learning_rate = 2e-4,
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir="outputs",
        report_to = "none",
    ),
)
````

We also use Unsloth's `train_on_completions` method to only train on the assistant outputs and ignore the loss on the user's inputs. This helps increase accuracy of finetunes!

````python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part = "<start_of_turn>assistant\n",  # CHANGED: "assistant" not "model"
)
````

### Inference
Let's run the model via Unsloth native inference! According to the `Gemma-3` team, the recommended settings for inference are `temperature = 1.0, top_p = 0.95, top_k = 64`

````python
messages = [
    {'role': 'system','content':dataset['conversations'][10][0]['content']},
    {"role" : 'user', 'content' : dataset['conversations'][10][1]['content']}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize = False,
    add_generation_prompt = True, # Must add for generation
).removeprefix('<bos>')

from transformers import TextStreamer
_ = model.generate(
    **tokenizer(text, return_tensors = "pt").to("cuda"),
    max_new_tokens = 125,
    temperature = 1, top_p = 0.95, top_k = 64,
    streamer = TextStreamer(tokenizer, skip_prompt = True),
)
````

````python
# ==== Colab test harness for a GGUF fine-tuned VCC model ====
# - If your files aren't in /content, you'll be prompted to upload them.
# - Loads GGUF with llama-cpp-python
# - Runs sanity probes
# - (Optional) Evaluates on your JSONL with exact-match + ROUGE-L

!pip -q install llama-cpp-python rouge-score

import os, json, random, math, sys
from pathlib import Path
from rouge_score import rouge_scorer
from llama_cpp import Llama

# ---- CONFIG (edit these if your filenames differ) ----
MODEL_PATH = "/content/gemma-3-270m-it.Q4_K_M.gguf"     # <-- your .gguf
DATA_PATH  = "/content/voice_commands_converted.jsonl"  # <-- your JSONL (optional)
SYSTEM = "Convert the following natural language command to the correct voice control command format."
STOP_TOKENS = ["<|end|>"]
CTX_TOKENS = 2048
N_THREADS = max(2, os.cpu_count() or 8)

# ---- Helper: upload files if missing ----
def ensure_file(path, label):
    p = Path(path)
    if p.exists():
        print(f"[OK] Found {label} at {path}")
        return str(p)
    from google.colab import files
    print(f"[UPLOAD] Please upload {label} now ‚Üí")
    uploaded = files.upload()
    if not uploaded:
        raise RuntimeError(f"No {label} uploaded.")
    # take the first uploaded name
    up_name = list(uploaded.keys())[0]
    print(f"[SAVED] {label} uploaded as: /content/{up_name}")
    return f"/content/{up_name}"

MODEL_PATH = ensure_file(MODEL_PATH, "GGUF model (.gguf)")
if DATA_PATH:  # optional
    if not Path(DATA_PATH).exists():
        try:
            DATA_PATH = ensure_file(DATA_PATH, "dataset (.jsonl)")
        except Exception:
            print("[WARN] Dataset not uploaded; will skip dataset eval.")
            DATA_PATH = ""

# ---- Load model (deterministic settings) ----
llm = Llama(
    model_path=MODEL_PATH,
    n_ctx=CTX_TOKENS,
    n_threads=N_THREADS,
    temperature=0.0,  # greedy
)

def infer_vcc(natural: str, max_new_tokens: int = 24) -> str:
    """Normalize a natural command using Gemma-3 chat tags + stop token."""
    prompt = (
        f"<|system|>\n{SYSTEM}<|end|>\n"
        f"<|user|>\n{natural}\n<|end|>\n"
        f"<|assistant|>\n"
    )
    out = llm(
        prompt,
        max_tokens=max_new_tokens,
        temperature=0.0,
        top_p=1.0,
        stop=STOP_TOKENS,
    )
    return out["choices"][0]["text"].strip()

# ---- Sanity probes ----
tests = [
    ("I was wondering if you could select 'tonight' | selection: ", "SELECT tonight"),
    ("lower the volume", "DECREASE VOLUME"),
    ("please open Spotify", "OPEN Spotify"),
    ("please replace today with tomorrow", "REPLACE today WITH tomorrow"),
    ("scroll down please", "SCROLL DOWN"),
]
print("=== Sanity checks ===")
ok = 0
for inp, exp in tests:
    pred = infer_vcc(inp)
    print(f"IN : {inp}\nOUT: {pred}\nEXP: {exp}\n---")
    ok += int(pred == exp)
print(f"Exact match (sanity): {ok}/{len(tests)}")

# ---- Dataset eval (optional) ----
if DATA_PATH and Path(DATA_PATH).exists():
    print("\n=== Dataset eval (random 100 examples) ===")
    data = []
    with open(DATA_PATH, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            try:
                ex = json.loads(line)
                # Expect your schema: task, input, expected_output, conversations
                if "input" in ex and "expected_output" in ex:
                    data.append(ex)
            except json.JSONDecodeError:
                pass

    if not data:
        print("[WARN] No valid examples loaded from dataset; skipping.")
    else:
        sample = random.sample(data, min(100, len(data)))
        scorer = rouge_scorer.RougeScorer(["rougeL"], use_stemmer=True)
        exact, rsum = 0, 0.0

        for ex in sample:
            inp = ex["input"].strip()
            gold = ex["expected_output"].strip()
            pred = infer_vcc(inp)

            if pred == gold:
                exact += 1
            rsum += scorer.score(gold, pred)["rougeL"].fmeasure

        n = len(sample)
        print(f"Exact match: {exact}/{n} = {exact/n:.3f}")
        print(f"ROUGE-L F1: {rsum/n:.3f}")
else:
    print("\n[INFO] DATA_PATH not provided/found. Skipping dataset eval.")

print("\nDone. If outputs are verbose or multi-line, double-check your chat tags and stop token (<|end|>).")
````
